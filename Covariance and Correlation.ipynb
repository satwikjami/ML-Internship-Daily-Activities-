# Normalization

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import pandas as pd
```

```python
import pandas as pd 
s=pd.Series([1000,2000,3000,4000,9000])
s
```

```python
s.min()
```

```python
s.max()
```

```python
sn=(s-s.min())/(s.max()-s.min())
sn
```

```python
sn=MinMaxScaler().fit_transform(s.values.reshape(-1,1))
print(sn)
```

```python
data= {
    "temp": [30,32,34,35,36],   
    "humidity":[40,39,34,28,27]
}
df=pd.DataFrame(data)
df
```

```python
norm_data=MinMaxScaler().fit_transform(df) # to normalize the code
n_df= pd.DataFrame(norm_data,columns=df.columns) # it gives the data in an array
norm_data
```

```python
n_df # it displays in a structured format
```
